```python
flights = [
    {"source": "Delhi", "destination": "Goa", "price": 4800},
    {"source": "Mumbai", "destination": "Goa", "price": 3500}
]

print(flights)


```

```python
def search_flight(source, destination):
    for f in flights:
        if f["source"] == source and f["destination"] == destination:
            return f
```

```python
search_flight("Delhi", "Goa")
```

```python
hotels = [
    {"city": "Goa", "name": "Sea View Resort", "price": 3000},
    {"city": "Goa", "name": "Beach Resort", "price": 4000}
]

print(hotels)
```

```python
def recommend_hotel(city):
    for h in hotels:
        if h["city"] == city:
            return h
```

```python
recommend_hotel("Goa")
```

```python
places = [
    {"city": "Goa", "place": "Baga Beach"},
    {"city": "Goa", "place": "Calangute Beach"},
    {"city": "Goa", "place": "Basilica of Bom Jesus"}
]

print(places)
```

```python
def get_places(city):
    result = []
    for p in places:
        if p["city"] == city:
            result.append(p["place"])
    return result
```

```python
get_places("Goa")
```

```python
def travel_planner(source, destination):
    flight = search_flight(source, destination)
    hotel = recommend_hotel(destination)
    place_list = get_places(destination)

    print("‚úà Flight Selected:")
    print(flight)
    print()

    print("üè® Hotel Recommended:")
    print(hotel)
    print()

    print("üìç Places to Visit:")
    for p in place_list:
        print("-", p)
```

```python
travel_planner("Delhi", "Goa")
```

```python
def estimate_budget(flight, hotel, days):
    flight_cost = flight["price"]
    hotel_cost = hotel["price"] * days   # üî• yahan fix
    food_cost = days * 800

    total = flight_cost + hotel_cost + food_cost

    return {
        "Flight Cost": flight_cost,
        "Hotel Cost": hotel_cost,
        "Food & Travel": food_cost,
        "Total Budget": total
    }
```

```python
flight = search_flight("Delhi", "Goa")
hotel = recommend_hotel("Goa")

estimate_budget(flight, hotel, 3)
```

```python
def final_travel_planner(source, destination, days):
    flight = search_flight(source, destination)
    hotel = recommend_hotel(destination)
    place_list = get_places(destination)
    budget = estimate_budget(flight, hotel, days)

    print("üåç FINAL TRAVEL PLAN")
    print("====================")

    print("\n‚úà Flight Selected")
    print(f"Airline: {flight.get('airline', 'N/A')}")
    print(f"From {flight['source']} to {flight['destination']}")
    print(f"Price: ‚Çπ{flight['price']}")

    print("\nüè® Hotel Recommended")
    print(f"Name: {hotel['name']}")
    print(f"City: {hotel['city']}")
    print(f"Price per night: ‚Çπ{hotel['price']}")

    print("\nüìç Places to Visit")
    for p in place_list:
        print("-", p)

    print("\nüí∞ Budget Breakdown")
    for k, v in budget.items():
        print(f"{k}: ‚Çπ{v}")
```

```python
final_travel_planner("Delhi", "Goa", 3)
```
